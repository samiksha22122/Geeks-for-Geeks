###create new column using for loop

```python
raw_Data = {'Voter_name': ['Geek1', 'Geek2', 'Geek3', 'Geek4',
                           'Geek5', 'Geek6', 'Geek7', 'Geek8'],
            'Voter_age': [15, 23, 25, 9, 67, 54, 42, np.nan]}

df = pd.DataFrame(raw_Data, columns = ['Voter_name', 'Voter_age'])
df
```

```python
df['Voter']=df['Voter_age'].apply(lambda x:'yes' if x>=18 else 'no')
df
```

```python
eligible=[]
for i in df['Voter_age']:
  if i>=18:
    eligible.append('Yes')
  elif i<18:
    eligible.append('No')
  else:
    eligible.append('Not Sure')
df['Voter']=eligible
df
```

```python
df = pd.DataFrame({'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/2011'],
				'Product':['Umbrella', 'Mattress', 'Badminton', 'Shuttle'],
				'Last Price':[1200, 1500, 1600, 352],
				'Updated Price':[1250, 1450, 1550, 400],
				'Discount':[10, 10, 10, 10]})
df
```

###Now we will check if the updated price is available or not. If not available then we will apply the discount of 10% on the ‘Last Price’ column to calculate the final price.

```python
# Check if the updated price is available or not
if 'Updated Price' in df.columns:
	df['Final cost'] = df['Updated Price'] - (df['Updated Price']*0.1)

else :
	df['Final cost'] = df['Last Price'] - (df['Last Price']*0.1)
df
```

To break up the string we will use Series.str.extract(pat, flags=0, expand=True) function. Here pat refers to the pattern that we want to search for.

```python
import pandas as pd

dict = {'movie_data':['The Godfather 1972 9.2',
					'Bird Box 2018 6.8',
					'Fight Club 1999 8.8'] }
df = pd.DataFrame(dict)
df
```

```python
# Extract name from the string
df['Name'] = df['movie_data'].str.extract('(\w*\s\w*)', expand=True)

# Extract year from the string
df['Year'] = df['movie_data'].str.extract('(\d\d\d\d)', expand=True)

# Extract rating from the string
df['Rating'] = df['movie_data'].str.extract('(\d\.\d)', expand=True)
df
```

##regex

```python
text = "hello world"
match = re.match(r"hello", text)
if match:
    print("Pattern found!")
else:
    print("No match.")
```

```python
text = "I love Python. Python is great!"
match = re.search(r"Python", text)
if match:
    print("Found:", match.group())
else:
    print("Not found.")
```

```python
text = "apple banana apple orange apple"
matches = re.findall(r"apple", text)
print(matches)
```

```python
text = "I love Python. Python is fun!"
pattern = r"Python"
replacement = "Java"

new_text = re.sub(pattern, replacement, text)
print(new_text)
```

```python
text = "word1,word2;word3 word4.word5"
pattern = r"[ ,;.]"  # Match space, comma, semicolon, or dot

words = re.split(pattern, text)
print(words)
```

```python
# Example 1: Any character except newline
text = "abc axc a7c ac"
print(re.findall(r'a.c', text))
```

```python
# Example 2: Start of a string
text = "hello world\nworld hello\nhello there"
re.findall(r'^hello+',text,re.MULTILINE)
```

```python
# Example 3: End of a string
text = "hello world\nworld hello\nthis is a world"
print(re.findall(r'world$', text, re.MULTILINE))
```

```python
# Example 4: 0 or more repetitions
text = "a ab abb abbb ac"
re.findall(r'ab*',text)
```

```python
# Example 5: 1 or more repetitions
text = "a ab abb abbb ac"
print(re.findall(r'ab+', text))
```

```python
# Example 6: 0 or 1 occurrence
text = "a ab abb ac"
print(re.findall(r'ab?', text))
```

```python
# Example 7: Exactly n times
text = "aa aaa aaaa"
print(re.findall(r'a{3}', text))
```

```python
# Example 8: At least n times
text = "a aa aaa aaaa"
print(re.findall(r'a{2,}', text))
```

```python
# Example 9: Between n and m times
text = "a aa aaa aaaa aaaaa"
print(re.findall(r'a{2,4}', text))
```

```python
# Example 10: Character set
text = "a e i o u b c"
print(re.findall(r'[aeiou]', text))
```

```python
# Example 15: Word character (a-z, A-Z, 0-9, _)
text = "A 1 _ @"
print(re.findall(r'\w', text))
```

```python
# Example 16: Non-word character
text = "A 1 _ @ !"
print(re.findall(r'\W', text))
```

```python
# Example 11: Digit (0-9)
text = "5 9 A @"
print(re.findall(r'\d', text))
```

```python
# Example 12: Non-digit
text = "5 9 A @"
print(re.findall(r'\D', text))
```

```python
# Example 13: Whitespace
text = "  \t A 1"
print(re.findall(r'\s', text))
```

```python
# Example 14: Non-whitespace
text = "  A 1 @"
print(re.findall(r'\S', text))
```

```python
text = "abc abcee aABC"
print(re.findall(r'(ab)+', text))
```
